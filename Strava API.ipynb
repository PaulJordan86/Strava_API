```python
import requests
import urllib3
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

auth_url = "https://www.strava.com/oauth/token"
activites_url = "https://www.strava.com/api/v3/athlete/activities"

payload = {
    'client_id': "106860",
    'client_secret': '2a3d94dfdf01b1a8700889325489040fb5e70541',
    'grant_type': 'authorization_code',
    'code': '96c3b573770f17b949bf86a6ab84d3c10c117941',
    'f': 'json'
}

print("Requesting Token...\n")
res = requests.post(auth_url, data=payload, verify=False)
access_token = res.json()['access_token']
print("Access Token = {}\n".format(access_token))

header = {'Authorization': 'Bearer ' + access_token}

request_page_num = 1
all_activities = []

while True:
    
    param = {'per_page': 200, 'page': request_page_num}
    my_dataset = requests.get(activites_url, headers=header, params=param).json() 
    print(len(my_dataset))
    if (len(my_dataset)) == 0:
       
        print("breaking out of while loop because no more activities to access")
        break
    
    if all_activities:
        print("all_activities is populated")
        all_activities.extend(my_dataset)
        
    else:
        print("all_activities is not populated")
        all_activities = my_dataset
    
    request_page_num += 1


```
